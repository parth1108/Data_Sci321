```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure(figsize=(15,5))
x = np.arange(30)
ax = fig.add_axes([0,0,1,1])
ax.bar(x+0.0,df['active'],color='green',ec='k',label='Active_Case',width=0.3)
ax.bar(x+0.2,df['confirmed'],color='red',ec='k',label='Confirm_Case',width=0.3)
ax.bar(x+0.4,df['deceased'],color='grey',ec='k',label='Death',width=0.3)
ax.set_xticks(x)
ax.set_xticklabels(df['district'],rotation=70)
plt.title("Covid-19 Guj",fontsize=25)
plt.xlabel('District')
plt.ylabel('Count')
plt.grid()
ax.legend()
plt.show()
```
